### **Bank Teller Crisis – Banking Transaction Management System**


**Objective**

To simulate the basic operations performed by a bank teller:

*  Check account balance
*  Deposit money
*  Withdraw money
*  Transfer money between accounts


**Accounts Used**
*  Checking Account
*  Savings Account


**Programming Concepts Demonstrated**
*  Python Functions
*  Variables
*  String Formatting (f-strings)
*  Conditionals
*  Business Logic
*  State Management

---



#### **1. Initial Balances**

```python
checking_balance = 0
savings_balance = 0

acc_error = 'Unsuccessful, please enter "checking" or "savings"'
```

Stores a reusable error message. Instead of writing the same message repeatedly, the program stores it once.

---



#### **2. Check Balance Function**

```python
def check_balance(account_type, checking_balance, savings_balance):

    if account_type == "savings":
        balance = savings_balance

    elif account_type == "checking":
        balance = checking_balance

    else:
        return acc_error

    balance_statement = (
        f"Your {account_type} balance is ${balance}"
    )

    return balance_statement
```

Returns the message to the caller.

---



#### **3. Deposit Function**

```python
def make_deposit(
    account_type,
    amount,
    checking_balance,
    savings_balance
):

    deposit_status = ""

    if amount > 0:

        if account_type == "savings":
            savings_balance += amount
            deposit_status = "successful"

        elif account_type == "checking":
            checking_balance += amount
            deposit_status = "successful"

        else:
            deposit_status = acc_error

    else:
        deposit_status = (
            "unsuccessful, amount must be greater than 0"
        )

    print(
        f"Deposit of ${amount} to your {account_type} account was {deposit_status}."
    )

    return savings_balance, checking_balance
```

#### **4. Withdrawal Function**

```python
def make_withdrawal(
    account_type,
    amount,
    checking_balance,
    savings_balance
):

    withdrawal_status = ""

    fail = (
        "unsuccessful, insufficient funds"
    )

    if account_type == "savings":

        if amount <= savings_balance:
            savings_balance -= amount
            withdrawal_status = "successful"

        else:
            withdrawal_status = fail

    elif account_type == "checking":

        if amount <= checking_balance:
            checking_balance -= amount
            withdrawal_status = "successful"

        else:
            withdrawal_status = fail

    else:
        withdrawal_status = acc_error

    print(
        f"Withdrawal of ${amount} from your {account_type} account was {withdrawal_status}."
    )

    return savings_balance, checking_balance
```

#### **5. Transfer Function**

```python
def acc_transfer(
    acc_from,
    acc_to,
    amount,
    checking_balance,
    savings_balance
):

    trans_status = ""

    if acc_from == "savings" and acc_to == "checking":

        if amount <= savings_balance:

            savings_balance -= amount
            checking_balance += amount

            trans_status = "successful"

        else:
            trans_status = "unsuccessful, insufficient savings balance"

    elif acc_from == "checking" and acc_to == "savings":

        if amount <= checking_balance:

            checking_balance -= amount
            savings_balance += amount

            trans_status = "successful"

        else:
            trans_status = "unsuccessful, insufficient checking balance"

    else:
        trans_status = acc_error

    print(
        f"Transfer of ${amount} from {acc_from} to {acc_to} was {trans_status}."
    )

    return savings_balance, checking_balance
```

#### **6. Main Program Execution**

```python
print("INITIAL BALANCES")
print(check_balance("checking", checking_balance, savings_balance))
print(check_balance("savings", checking_balance, savings_balance))

print("\n-----------------------------")
print("DEPOSIT TRANSACTIONS")
print("-----------------------------")

savings_balance, checking_balance = make_deposit(
    "savings",
    10,
    checking_balance,
    savings_balance
)

savings_balance, checking_balance = make_deposit(
    "checking",
    200,
    checking_balance,
    savings_balance
)

print(check_balance("checking", checking_balance, savings_balance))
print(check_balance("savings", checking_balance, savings_balance))

print("\n-----------------------------")
print("WITHDRAWAL TRANSACTIONS")
print("-----------------------------")

savings_balance, checking_balance = make_withdrawal(
    "savings",
    11,
    checking_balance,
    savings_balance
)

savings_balance, checking_balance = make_withdrawal(
    "checking",
    170,
    checking_balance,
    savings_balance
)

print(check_balance("checking", checking_balance, savings_balance))
print(check_balance("savings", checking_balance, savings_balance))

print("\n-----------------------------")
print("TRANSFER TRANSACTIONS")
print("-----------------------------")

savings_balance, checking_balance = acc_transfer(
    "checking",
    "savings",
    40,
    checking_balance,
    savings_balance
)

savings_balance, checking_balance = acc_transfer(
    "savings",
    "checking",
    5,
    checking_balance,
    savings_balance
)

print("\nFINAL BALANCES")
print(check_balance("checking", checking_balance, savings_balance))
print(check_balance("savings", checking_balance, savings_balance))
```
